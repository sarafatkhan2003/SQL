**WHERE Clause â€“ Filtering Data (CORE SKILL)**

```python
# Basic Filter

SELECT * FROM employees
WHERE salary > 50000;
```

```python
# Filter with Text

SELECT * FROM employees
WHERE department = 'IT';
```

Text values must be in single quotes

```python
# Multiple Conditions

SELECT * FROM employees
WHERE department = 'IT'
AND salary > 50000;
```

```python
SELECT * FROM employees
WHERE city = 'Delhi'
OR city = 'Mumbai'
```

ORDER BY â€“ Sorting Results

```python
# Default (Ascending)

SELECT * FROM employees
ORDER BY salary;
```

```python
# Descending

SELECT * FROM employees
ORDER BY salary DESC;
```

ORDER BY always comes after WHERE

LIMIT â€“ Control Output Size

```python
SELECT * FROM employees
ORDER BY salary DESC
LIMIT 3;
```

Used for:

- Top N records
- Sampling data
- Performance testing

SQL Query Execution Order (INTERVIEW GOLD)

This is CRITICAL.

Even though you write:

```python
SELECT
FROM
WHERE
ORDER BY
LIMIT
```

SQL Executes in This Order:

1ï¸âƒ£ FROM

2ï¸âƒ£ WHERE

3ï¸âƒ£ SELECT

4ï¸âƒ£ ORDER BY

5ï¸âƒ£ LIMIT

ðŸ“Œ SQL reads FROM first, not SELECT.

```python
Example Breakdown
SELECT name, salary
FROM employees
WHERE salary > 50000
ORDER BY salary DESC
LIMIT 2;
```

Execution:

Read employees table

Filter salary > 50,000

Select name & salary columns

Sort by salary (DESC)

Return top 2 rows
