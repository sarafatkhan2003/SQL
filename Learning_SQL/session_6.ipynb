 **ðŸ”¹ WHERE vs HAVING (Concept)**

| Clause | Filters                    |
| ------ | -------------------------- |
| WHERE  | Rows (before grouping)     |
| HAVING | Groups (after aggregation) |

ðŸ“Œ WHERE â†’ before GROUP BY

**âŒ WRONG (Very Common Mistake)**

```sql
SELECT department, AVG(salary)
FROM employees
WHERE AVG(salary) > 60000   -- âŒ INVALID
GROUP BY department;
```

Why?

* WHERE runs **before** AVG exists.

```sql
SELECT department, AVG(salary)
FROM employees
GROUP BY department
HAVING AVG(salary) > 60000;
```

**4ï¸âƒ£ WHERE + GROUP BY + HAVING (FULL POWER)**

**Example:**

```sql
SELECT department, COUNT(*)
FROM employees
WHERE city = 'Delhi'
GROUP BY department
HAVING COUNT(*) > 1;
```

1. FROM

4. HAVING (filter groups)
