**ðŸ”¹ Why ORDER BY Matters**

**ðŸ”¸ Basic ORDER BY**

```sql
SELECT * FROM employees
ORDER BY salary;
```

* Default = **ASC** (ascending)

**ðŸ”¸ DESC (Descending Order)**

```sql
SELECT * FROM employees
ORDER BY salary DESC;
```

ðŸ“Œ Used for:

* Top salaries

**ðŸ”¸ ORDER BY Multiple Columns**

```sql
SELECT * FROM employees
ORDER BY department ASC, salary DESC;
```

 **2ï¸âƒ£ LIMIT (Top / Bottom N Records)**

 **ðŸ”¹ Why LIMIT Matters**

---

 **ðŸ”¸ Basic LIMIT**

```sql
SELECT * FROM employees
LIMIT 3;
```

ðŸ“Œ Returns **any 3 rows** (not meaningful without ORDER BY).

 **ðŸ”¸ LIMIT with ORDER BY (Correct Usage)**

```sql
SELECT * FROM employees
ORDER BY salary DESC
LIMIT 2;
```

ðŸ“Œ **Always pair LIMIT with ORDER BY**.

 **3ï¸âƒ£ Aliases with AS (Very Important)**

**ðŸ”¹ What Is an Alias?**

An alias:

* Renames a column or table
* Improves readability
* Makes results interview-presentable

---

 **ðŸ”¸ Alias for Columns**

```sql
SELECT AVG(salary) AS avg_salary
FROM employees;
```

Output column name becomes: `avg_salary`.

---

 **ðŸ”¸ Alias with GROUP BY**

```sql
SELECT department, COUNT(*) AS total_employees
FROM employees
GROUP BY department;
```

ðŸ“Œ Much clearer than `count`.

---

 **ðŸ”¸ Alias for Tables (Used in Joins Later)**

```sql
SELECT e.name, e.salary
FROM employees AS e;
```

ðŸ“Œ Makes queries shorter and cleaner.

---

 **4ï¸âƒ£ ORDER BY Using Aliases (INTERVIEW FAVORITE)**

 **Example**

```sql
SELECT department, AVG(salary) AS avg_salary
FROM employees
GROUP BY department
ORDER BY avg_salary DESC;
```

ðŸ“Œ This works because:

* ORDER BY runs **after SELECT**
* Alias already exists

 **5ï¸âƒ£ ORDER BY Column Position (Use Carefully)**

**6ï¸âƒ£ Real-World Analyst Queries (Putting It All Together)**

 **Example 1: Top-paying Departments**

```sql
SELECT department, AVG(salary) AS avg_salary
FROM employees
GROUP BY department
HAVING AVG(salary) > 50000
ORDER BY avg_salary DESC
LIMIT 2;
```

---

 **Example 2: City-wise Employee Ranking**

```sql
SELECT city, COUNT(*) AS employee_count
FROM employees
GROUP BY city
ORDER BY employee_count DESC;
```
